```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
```

```python
tokenizer = AutoTokenizer.from_pretrained("oliverguhr/fullstop-punctuation-multilang-large")
```

```python
model = AutoModelForTokenClassification.from_pretrained("oliverguhr/fullstop-punctuation-multilang-large")
```

```python
pun = pipeline('ner', model=model, tokenizer=tokenizer)
```

```python
text = "Hey My name is Sai Prasad I am 20 years old"
```

```python
output_json = pun(text)
```

```python
output_json
```

```python
s = ''

for n in output_json:
  result = n['word'].replace('▁',' ') + n['entity'].replace('0','')
  s += result
```

```python
s
```

```python
from spacy import displacy
```

```python
empty_ner = []
for n in output_json:
  if n['entity'] != '0':
    temp = [n['start'],n['end'],n['entity']]
    empty_ner.append(temp)
```

```python
empty_ner
```

```python
colors = {",": "linear-gradient(90deg, #EFED65, #EFED65)",
          ".": "linear-gradient(90deg, #B6FF7C, #B6FF7C)",
          "?": "linear-gradient(90deg, #FF7C7C, #FF7C7C)",
          ":": "linear-gradient(90deg, #FF7C7C, #FF7C7C)",
          "-": "linear-gradient(90deg, #FF7C7C, #FF7C7C)"}
options = {"ents": [",",".","?",":",'-'], "colors": colors}
```

```python
[{"start": x[0], "end": x[1], "label": x[2]} for x in empty_ner]
```

```python
ex = [{"text": text,
       "ents": [{"start": x[0], "end": x[1], "label": x[2]} for x in empty_ner],
       "title": "MINI PROJECT"}]
```

```python
html = displacy.render(ex, style="ent", manual=True, options=options)
```
